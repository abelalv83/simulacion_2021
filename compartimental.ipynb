# Modelos compartimentales 

Los modelos compartimentalesson una técnica modelizar matemáticamente enfermedades infeccionsas. Para ello se divide la población en pequeños grupos con caracteristicas similares epidemologicamente. Uno de los primeros autores que trabajaron con estos modelos son [Kermack y McKendrick en 1927](https://royalsocietypublishing.org/doi/10.1098/rspa.1927.0118) el modelo simple es  SIR, el cual divide la población en tres grupos: subsetible, infectado y recuperado. **Los susceptibles** son aquellos individuos sanos que pueden contagiarse de la enfermedad en el un tiempo determinado. **Los individuos infectados** son aquellos que que contrajeron la enfermedad y **los recuperados** son aquellos individuos que ya contrajeron la enfermedad y están sanos

Para analizar este modelo, podemos usar ecuaciones diferenciales, en el cual deteminamos que existen tres variables $S$, $I$ y $R$, las cuales pueden variar en el tiempo, de esta forma podemos determinar, las ecuaciones 
$$\frac{dS}{dt}.$$
como la tasa de variacíon de la población susceptible en un tiempo determinado, usando el principio de conservación tendriamos que esta tasa seria igual a la que ingresa al compartimiento de susceptible (que en este ejemplo es cero) menos la población que deja el compartimiento y pasa al compartimiento de población infectada, la cual la nombraremos como $\mu$ , así 
$$\frac{dS}{dt}=-\mu.$$ 
Note que la población infectada podría ser calculada como proporcional a los encuentros entre la población infectados con la población susceptible, asi tendriamos 
$$\mu=\kappa I S,$$ 
así
$$\frac{dS}{dt}=-\kappa I S.$$ 
De la misma forma podriamos análizar la tasa de  población infectada

$$\frac{dI}{dt}=\kappa IS -\sigma I,$$
aqui $\sigma$ significa la tasa de la población que se recupera en el tiempo $t$.
y finalmente, la tasa de recuperados podría ser  igual a 
$$ \frac{d R}{dt}=\sigma I.$$
