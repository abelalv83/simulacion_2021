#  Análisis dimensional
Es un método por el cual nosotros deducimos  información de un fenómeno desde 
una simple premisa: el fenómeno puede ser descrito por una ecuación 
dimensionalmente correcta entre ciertas variables.
 
 * Cada término separadamente debe ser dimensionalmente consistente.
 * Las unidades deben ser consistentes

tiene el mismo conjunto de dimensiones. Entonces, tomando esto en su totalidad

## The Buckingham Pi Theorem for Dimensional Analysis
  Una ecuación dimensional homogeneiza que involucra $n$ variables en $m$
  variables fundamentales o primarias puede ser reducido a una simple relación 
  entre $n-m$ producto de dimensiones independientes.
 \end{theorem}
 Si la notación $\Pi$ se usa para representar un termino dimensional, entonces 
el teorema de Buckingham puede escribirse
$$\Pi_1=\Phi(\Pi_2,\Pi_3,...\Pi_{n-m})$$
o equivalente 
$$\Phi(\Pi_1,\Pi_2,\Pi_3,...\Pi_{n-m})=0$$
Esta ecuación indica que  para un problema con $n$ variables derivadas y $m$
variables primarias requieren $n - m$ grupos adimensionales
para correlación de todas sus variables.

# Principios de conservación y balance
\begin{equation}
 \frac{dQ(t)}{dt}=q_{in}(t)+g(t)-q_{out}(t)-c(t).
\end{equation}
$Q(t)$ representa el flujo de las físicas propiedades, $q_{in}$ y 
$q_{out}$ representan las tasa de $Q(t)$ que entra y sale, respectivamente en 
la frontera. $g(t)$ es la tasa para la cual $Q$ es generado dentro del dominio.
$c(t)$ es la tasa para cual $Q$  es consumido dentro del dominio.
